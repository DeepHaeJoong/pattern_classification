### 1.1 Consider basic gradient descent (Algorithm 1) and Newton’s algorithm (Algorithm 2) applied to the data in the tables.

> **criterion function**
$$
J_{p}(\mathbf{a})=\sum_{\mathbf{y} \in \mathcal{Y}}(-\mathbf{a}^T\mathbf{y}) \tag{16}\label{16}
$$


> **simple gradient descent**
$$
\mathbf{a}(k+1)=\mathbf{a}(k)-\eta(k) \nabla J(\mathbf{a}(k))
$$

> **Newton's algorithm**
$$
\mathbf{a}(k+1)=\mathbf{a}(k)-\mathbf{H}^{-1} \nabla J
$$

#### (a) $\omega_1$과 $\omega_2$를 구분하기 위해 2-차원 데이터에 적용해보자. 즉, simple gradient descent을 사용하여 criterion function을 반복 횟수 함수로서 그려보자. (단, $\eta(k) = 0.01$ 사용)

> **`update rule`**
$$
\mathbf{a}(k+1)=\mathbf{a}(k)+\eta(k) \sum_{\mathbf{y} \in \mathcal{Y}} \mathbf{y}
$$

```python
# simple gradient descent (algorithm 1)

learning_rate = 0.01
theta = 0.001

# initialize "w_vector"
#w_vector = np.array([[2], [3]])
w_vector = np.array([[1], [-3]])
J = 0
epoch_num = 2000

for k in range(epoch_num):
    temp = 0
    loss_gradient_sum = 0
    for i,j in enumerate(ex_1_data.values):
        if j[2] == 0:
            temp = w_vector.T @ np.array([j[0],j[1]])
            #print("J({1}) : a^T y_{0} = {2}".format(i, j[2], temp))

        else :
            temp = - w_vector.T @ np.array([j[0],j[1]])
            #print("J({1}) : a^T y_{0} = {2}".format(i, j[2], temp))
        
        if temp < 0:
            # print('error')
            loss_gradient_sum += np.array([[j[0]], [j[1]]])
            
    w_vector = w_vector + learning_rate*(loss_gradient_sum)
    
    if np.absolute(learning_rate*np.linalg.norm(loss_gradient_sum)) < theta:
        print('stop !')
        break


w_vector
```
